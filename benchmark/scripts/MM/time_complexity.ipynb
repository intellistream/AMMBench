```python
import matplotlib.pyplot as plt
```

```python
n1 = 10000
n2 = 10000
ms = [1000, 5000, 10000, 50000]
```

```python
ss=0.1

def get_time_complexity_of_mm(n1, n2, m):
    return n1*n2*m

def get_time_complexity_of_crs(n1, n2, m):
    return ss*n1*n2*m

def get_time_complexity_of_smppca(n1, n2, m):
    return (n1+n2)*(0.1*m**2+1.2*m) + 4*n1*n2
```

```python
get_time_complexity_of_mm(10000, 10000, 20000)/get_time_complexity_of_smppca(10000, 10000, 20000)
# should be 5.5
```

```python
time_complexity_of_mm = [1/get_time_complexity_of_mm(n1, n2, m) for m in ms]
time_complexity_of_crs = [1/get_time_complexity_of_crs(n1, n2, m) for m in ms]
time_complexity_of_smppca = [1/get_time_complexity_of_smppca(n1, n2, m) for m in ms]
time_complexity_of_smppca
```

```python
# Plotting
plt.plot(ms, time_complexity_of_mm, color='green', label='mm')
plt.plot(ms, time_complexity_of_crs, color='blue', label='crs')
plt.plot(ms, time_complexity_of_smppca, color='orange', label='smppca')

# Set labels and title
plt.xlabel('m')
plt.ylabel('elasped time')
plt.title('Plot with Multiple Lines')

# Set legend
plt.legend()

# Show the plot
plt.show()
```
